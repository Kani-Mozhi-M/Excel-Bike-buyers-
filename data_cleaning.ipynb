1. Missing Values:

Question: Are there any missing values in the dataset, and if so, how should they be handled for each indicator?

```python
avg_age = data['Age'].mean()
avg_age
```

```python
data['Age'] = data.Age.fillna(avg_age)
```

```python
data['Age'] = data.Age.fillna(avg_age)
data['Age'] = data.Age.round(decimals=0)

data['Salary'] = data['Salary'].str.replace('$', '')

data['Salary'] = data.Salary.str.replace("k","000")

data['Easy Apply'] = data['Easy Apply'].replace('-1',"FALSE")


data['Rating'] = data.Rating.round(decimals = 1).abs().replace('Nan','0')
data = data.dropna(subset=['Rating'])


data['Location'] = data['Location'].str.split(",", expand=True) [0].str.replace(r'\sAus', '').str.replace(r'\sIn', '')
```

3. Outliers:

Question: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?

```python
Outlier = ['Age','Rating']
for i in Outlier:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[i] < lower_bound) | (data[i] > upper_bound)]

    print(f"Potential Outliers in {i}: ")
    print(outliers)
```
